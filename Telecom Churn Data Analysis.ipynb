```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("Customer churn.csv")
print(data)

data.head(10)
```

```python
data.info()
```

```python
data["TotalCharges"] = data["TotalCharges"].replace(" ","0") #replacing blank spaces with 0 
data["TotalCharges"] = data["TotalCharges"].astype("float")  #changing the datatype from object to float
data.info()
```

```python
data.isnull().sum()
```

```python
data.describe()
```

```python
data["customerID"].duplicated().sum()
```

```python
#function for converting 0 and 1 to yes and no in SeniorCitizen Column
def convert(val):
    if val == 1:
        return "yes"
    else:
        return "no"

data["SeniorCitizen"] = data["SeniorCitizen"].apply(convert)
data.head()
```

```python
#count of people who have churned out using count plot
plt.figure(figsize=[8,4]) 
plt.subplot(1,2,1)
val = sns.countplot(data = data , x = "Churn")
val.bar_label(val.containers[0])
plt.title("Count of People by Churn")

#percentage of people who have churned out using pie chart
plt.subplot(1,2,2)
grp = data.groupby("Churn").agg({"Churn":"count"})
plt.pie(grp["Churn"],labels = grp.index , autopct = "%.2f%%")
plt.title("Percentage of People by Churn")
plt.show()
```

#from the above countplot we got to know that 1869 customers churned out and with pie chart we got to know that 26.54% customers churned out
#Let's find the the reason behind it 

```python
#checking the number of male and female customers with how many of them churned out

plt.figure(figsize = (5,4))
val = sns.countplot(data = data , x = "gender" , hue = "Churn")
val.bar_label(val.containers[0])
val.bar_label(val.containers[1])
plt.title("Churn by Gender")
plt.show()
```

```python
#checking the number of senior citizen with how many of them churned out

plt.figure(figsize = (5,4))
val = sns.countplot(data = data , x = "SeniorCitizen",hue = "Churn")
val.bar_label(val.containers[0])
val.bar_label(val.containers[1])
plt.title("Count of Senior Citizen by Churn")
plt.show()
```

#After the comparision got to know that large number of senior citizens have churned out

```python
#Using a histogram plot to analyze the distribution of customer tenure, showing how long customers have been using our services
plt.figure(figsize = (8,4))
sns.histplot(data = data , x = "tenure",bins = 36 , hue = "Churn")
plt.show()
```

#This shows that maximum customer churned out in the initial month of our service , and maximum customers stayed have been using our services from long time

```python
#using count plot to count the customers based on their contract type and to analyze how many of them have churned
plt.figure(figsize = (5,4))
val = sns.countplot(data = data , x = "Contract" , hue = "Churn")
val.bar_label(val.containers[0])
val.bar_label(val.containers[1])
plt.title("Count of Customers Churn by Contract")
plt.show()
```

#Maximun number of customers who have churned are on month-to-month contract

```python
#Analyzing count of cutomers using various services and there churn rate
df = pd.DataFrame(data)
columns = ['PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity', 
           'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies']

# Set up the subplot grid (3 rows x 3 columns for 9 features)
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(18, 12))
fig.suptitle('Count Charts for Service Usage Analysis', fontsize=18)

# Flatten the axes array for easier iteration
axes = axes.flatten()

# Loop through each column and create a countplot
for i, column in enumerate(columns):
    val = sns.countplot(x=df[column], ax=axes[i],hue = df["Churn"])
    axes[i].set_title(f'{column} Counts')
    axes[i].set_xlabel('')
    axes[i].set_ylabel('Count')
    
#Adding labels to each bar
    for container in val.containers:
         val.bar_label(container)
        
# Adjust layout for better fit
plt.tight_layout()
plt.subplots_adjust(top=0.92)  # Leave space for the main title
plt.show()

```

#Most customers who stayed with the company have services like PhoneService, Fiber optic Internet, and extra features like OnlineSecurity, OnlineBackup, and DeviceProtection. Customers who didn’t have these services were more likely to leave.

```python
#Anayzing customer churn rate on the basis of payment method using pie and count plots

#Pie chart to demonstrate percentage of customers using various payments methods
plt.figure(figsize = (14,4)) #(width,height)
plt.subplot(1,2,1)
grp = df.groupby("PaymentMethod").agg({"PaymentMethod":"count"})
plt.pie(grp["PaymentMethod"] , labels= grp.index, autopct = "%.2f%%",textprops={'fontsize': 8},shadow = True)

#Count Plot to show count of customers using various PaymentMethods and their churn rate
plt.subplot(1,2,2)
val = sns.countplot(data = df , x = "PaymentMethod" , hue = "Churn",  )
val.bar_label(val.containers[0],fontsize = 8)
val.bar_label(val.containers[1],fontsize = 8)
val.set_xticklabels(val.get_xticklabels(), fontsize=7)
# plt.xticks(rotation = 40)
plt.legend(bbox_to_anchor = [1.2,1.0])
plt.show()
```

#The most popular payment method among customers is "Electronic check," followed by "Mailed check." A higher number of customers who use "Electronic check" have churned compared to other payment methods, where the churn rate is relatively lower.

```python
# Box plot for Monthly Charges and Total Charges analysis
plt.figure(figsize= (10, 4))

# Monthly Charges
plt.subplot(1, 2, 1)
sns.boxplot(data=df, y='MonthlyCharges',hue="Churn")
plt.title('Box Plot: Monthly Charges')
plt.ylabel('Monthly Charges')

# Total Charges
plt.subplot(1, 2, 2)
sns.boxplot(data=df, y='TotalCharges',hue="Churn")
plt.title('Box Plot: Total Charges')
plt.ylabel('Total Charges')

plt.subplots_adjust(wspace=0.5)
# plt.tight_layout()
plt.show()
```

#The above analysis shows that the maximum monthly charges lies between 35/- to 90/- , while maximum monthly charges lies between 800/- to 3500/- and the median value is approx 70/- and 1600/- respectively for both the plots
